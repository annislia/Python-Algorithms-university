1. Дан список. Выбрать из списка все нечетные числа и упорядочить их по убыванию.

```python
n=[1,2,3,5,6,7,9,10]
nn=[]
for i in n:
    if i%2!=0:
        nn.append(i)
print(sorted(nn, reverse=True))
    
```

```python
m = int(input('кольчество строк'))
n = int(input('количество столбцов'))
mat = []
print("Введите элементы матрицы построчно:")
for i in range(m):
    el = list(map(int, input().split()))
    mat.append(el)

print(mat)

for i in range(m//2):
    for j in range(n//2):
        i2 = i + m // 2
        j2 = j + n // 2
        mat[i][j], mat[i2][j2] = mat[i2][j2], mat[i][j]
        
        
print(mat)
```
