Примените структуру данных множества для решения следующих задач:</br>а) Используя множества, вывести все уникальные строчные русские гласные буквы из заданного текста в алфавитном порядке.</br>б) В базе данных содержится информация о продуктах в интернет-магазине. Каждый продукт имеет уникальный идентификационный номер и список категорий, к которым он относится. Необходимо написать программу, которая будет определять, в каких категориях есть только один продукт.

```python
t = "Пример текста для проверки уникальных гласных букв"
v = set("аеёиоуыэюя")
r = sorted(set(t.lower()) & v)
print("Уникальные гласные:", ''.join(r))

p = {
    1: ['эл', 'тел'],
    2: ['эл', 'ком'],
    3: ['кни'],
    4: ['акс', 'тел']
}

# Собираем все категории
c = []
for x in p.values():
    c.extend(x)

# Подсчитываем продукты в категориях
from collections import Counter
cnt = Counter(c)

# Категории с одним продуктом
res = [cat for cat, n in cnt.items() if n == 1]
print("Категории с одним продуктом:", res)
```

Примените структуру данных множества для решения следующих задач:</br>а) Используя множества, определить, каких букв больше в заданном тексте — гласных или согласных.</br>б) В базе данных содержится информация о заказах в интернет-магазине. Каждый заказ имеет уникальный номер и список товаров, которые были заказаны. Необходимо написать программу, которая будет определять, какие товары были заказаны только в определенный день.

```python
t = "Пример текста для сравнения букв"  # заданный текст
v = set("аеёиоуыэюя")  # гласные
c = set("бвгджзйклмнпрстфхцчшщ")  # согласные

t_low = t.lower()  # текст в нижнем регистре
v_cnt = len([x for x in t_low if x in v])  # счетчик гласных
c_cnt = len([x for x in t_low if x in c])  # счетчик согласных

if v_cnt > c_cnt:
    print("Гласных больше")
elif c_cnt > v_cnt:
    print("Согласных больше")
else:
    print("Поровну")


orders = {
    "день1": ["товар1", "товар2", "товар3"],
    "день2": ["товар2", "товар4", "товар5"],
    "день3": ["товар1", "товар6"]
}

# Собираем все товары и дни когда они заказывались
prod_days = {}
for day, prods in orders.items():
    for p in prods:
        if p not in prod_days:
            prod_days[p] = set()
        prod_days[p].add(day)

# Товары которые были только в один день
res = [p for p, days in prod_days.items() if len(days) == 1]
print("Товары только за один день:", res)
```

Каждый студент имеет уникальный идентификационный номер и список предметов, которые он изучает. Необходимо написать программу, которая будет определять, какие предметы изучают только студенты обоих университетов.

```python
A = {1, 2, 3, 4}
B = {3, 4, 5, 6}
res = len(A & B) > 0  # есть ли пересечение
print(res)

uni1 = {
    1: ["мат", "физ", "хим"],
    2: ["мат", "био", "инф"],
    3: ["физ", "хим"]
}

uni2 = {
    1: ["мат", "инф", "экон"],
    2: ["био", "хим", "физ"],
    3: ["инф", "мат"]
}

# Собираем все предметы каждого университета
sub1 = set()
for subs in uni1.values():
    sub1.update(subs)

sub2 = set()
for subs in uni2.values():
    sub2.update(subs)

# Предметы которые есть в обоих университетах
res = sub1 & sub2
print("Предметы в обоих университетах:", sorted(res))
```
