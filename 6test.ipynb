–î–∞–Ω —Å–ø–∏—Å–æ–∫, –≤–∫–ª—é—á–∞—é—â–∏–π  ùëÅ  —ç–ª–µ–º–µ–Ω—Ç–æ–≤, –∏ —Ü–µ–ª—ã–µ —á–∏—Å–ª–∞  ùêæ  –∏  ùêø  ( 1‚â§ùêæ‚â§ùêø‚â§ùëÅ ). –ù–∞–π—Ç–∏ —Å—Ä–µ–¥–Ω–µ–µ –∞—Ä–∏—Ñ–º–µ—Ç–∏—á–µ—Å–∫–æ–µ —ç–ª–µ–º–µ–Ω—Ç–æ–≤ —Å–ø–∏—Å–∫–∞ —Å –ø–æ—Ä—è–¥–∫–æ–≤—ã–º–∏ –Ω–æ–º–µ—Ä–∞–º–∏ –æ—Ç  ùêæ  –¥–æ  ùêø –≤–∫–ª—é—á–∏—Ç–µ–ª—å–Ω–æ.

```python
n=[1,2,3,4,5,6,7,8]
k,l= map(int, input().split())
print(sum(n[k:l+1])/len(n[k:l+1]))
```

–î–∞–Ω —Å–ø–∏—Å–æ–∫, –≤–∫–ª—é—á–∞—é—â–∏–π  ùëÅ  —ç–ª–µ–º–µ–Ω—Ç–æ–≤, –∏ —Ü–µ–ª—ã–µ —á–∏—Å–ª–∞  ùêæ  –∏  ùêø  ( 1<ùêæ‚â§ùêø‚â§ùëÅ ). –ù–∞–π—Ç–∏ —Å—É–º–º—É –≤—Å–µ—Ö —ç–ª–µ–º–µ–Ω—Ç–æ–≤ —Å–ø–∏—Å–∫–∞, –∫—Ä–æ–º–µ —ç–ª–µ–º–µ–Ω—Ç–æ–≤ —Å –ø–æ—Ä—è–¥–∫–æ–≤—ã–º–∏ –Ω–æ–º–µ—Ä–∞–º–∏ –æ—Ç  ùêæ  –¥–æ  ùêø –≤–∫–ª—é—á–∏—Ç–µ–ª—å–Ω–æ.

```python
n=[1,2,3,4,5,6,7,8]
k,l= map(int, input().split())
nn=n[k:l+1]
print(sum(n)-sum(nn))
```

–î–∞–Ω —Å–ø–∏—Å–æ–∫, –≤–∫–ª—é—á–∞—é—â–∏–π  ùëÅ  —ç–ª–µ–º–µ–Ω—Ç–æ–≤, –∏ —Ü–µ–ª—ã–µ —á–∏—Å–ª–∞  ùêæ  –∏  ùêø  ( 1<ùêæ‚â§ùêø‚â§ùëÅ ). –ù–∞–π—Ç–∏ —Å—Ä–µ–¥–Ω–µ–µ –∞—Ä–∏—Ñ–º–µ—Ç–∏—á–µ—Å–∫–æ–µ –≤—Å–µ—Ö —ç–ª–µ–º–µ–Ω—Ç–æ–≤ —Å–ø–∏—Å–∫–∞, –∫—Ä–æ–º–µ —ç–ª–µ–º–µ–Ω—Ç–æ–≤ —Å –ø–æ—Ä—è–¥–∫–æ–≤—ã–º–∏ –Ω–æ–º–µ—Ä–∞–º–∏ –æ—Ç  ùêæ  –¥–æ  ùêø  –≤–∫–ª—é—á–∏—Ç–µ–ª—å–Ω–æ.

```python
n=[1,2,3,4,5,6,7,8]
k,l= map(int, input().split())
nn=n[k:l+1]
s=sum(n)-sum(nn)
k=len(n)-len(nn)
print(s/k)
```

–î–∞–Ω —Å–ø–∏—Å–æ–∫. –í—ã–≤–µ—Å—Ç–∏ –≤–Ω–∞—á–∞–ª–µ –≤—Å–µ —Å–æ–¥–µ—Ä–∂–∞—â–∏–µ—Å—è –≤ –¥–∞–Ω–Ω–æ–º —Å–ø–∏—Å–∫–µ —á–µ—Ç–Ω—ã–µ —á–∏—Å–ª–∞ –≤ –ø–æ—Ä—è–¥–∫–µ –≤–æ–∑—Ä–∞—Å—Ç–∞–Ω–∏—è –∏—Ö –∏–Ω–¥–µ–∫—Å–æ–≤, –∞ –∑–∞—Ç–µ–º ‚Äî –≤—Å–µ –Ω–µ—á–µ—Ç–Ω—ã–µ —á–∏—Å–ª–∞ –≤ –ø–æ—Ä—è–¥–∫–µ —É–±—ã–≤–∞–Ω–∏—è –∏—Ö –∏–Ω–¥–µ–∫—Å–æ–≤.

```python
n=[1,2,3,4,5,6,7,8]
c=[x for x in n if x%2==0]
nc=[x for x in n if x%2!=0] [::-1]
print(c, nc, sep='\n')
```

–î–∞–Ω —Å–ø–∏—Å–æ–∫. –í—ã–≤–µ—Å—Ç–∏ –≤–Ω–∞—á–∞–ª–µ –µ–≥–æ —ç–ª–µ–º–µ–Ω—Ç—ã —Å —á–µ—Ç–Ω—ã–º–∏ –Ω–æ–º–µ—Ä–∞–º–∏ (–≤ –ø–æ—Ä—è–¥–∫–µ –≤–æ–∑—Ä–∞—Å—Ç–∞–Ω–∏—è –Ω–æ–º–µ—Ä–æ–≤), –∞ –∑–∞—Ç–µ–º ‚Äî —ç–ª–µ–º–µ–Ω—Ç—ã —Å –Ω–µ—á–µ—Ç–Ω—ã–º–∏ –Ω–æ–º–µ—Ä–∞–º–∏ (—Ç–∞–∫–∂–µ –≤ –ø–æ—Ä—è–¥–∫–µ –≤–æ–∑—Ä–∞—Å—Ç–∞–Ω–∏—è –Ω–æ–º–µ—Ä–æ–≤).

```python
n=[1,2,3,4,5,6,7,8]
c=[x for x in n if x%2==0]
nc=[x for x in n if x%2!=0] 
print(c, nc, sep='\n')
```

–î–∞–Ω —Å–ø–∏—Å–æ–∫. –ù–∞–π—Ç–∏ –ø—Ä–æ–∏–∑–≤–µ–¥–µ–Ω–∏–µ –≤—Å–µ—Ö —ç–ª–µ–º–µ–Ω—Ç–æ–≤ —Å–ø–∏—Å–∫–∞, —Ä–∞—Å–ø–æ–ª–æ–∂–µ–Ω–Ω—ã—Ö —Å–ø—Ä–∞–≤–∞ –æ—Ç –º–∞–∫—Å–∏–º–∞–ª—å–Ω–æ–≥–æ —ç–ª–µ–º–µ–Ω—Ç–∞–º–∏, –Ω–µ –≤–∫–ª—é—á–∞—è –º–∞–∫—Å–∏–º–∞–ª—å–Ω—ã–π —ç–ª–µ–º–µ–Ω—Ç.

```python
n=[1,2,3,4,5,6,7,8,4,3]
i=n.index(max(n))
p=1
nn=n[i+1:]
for x in nn:
    p*=x
print(p)
```

–î–∞–Ω —Å–ø–∏—Å–æ–∫. –ù–∞–π—Ç–∏ —Å—É–º–º—É –≤—Å–µ—Ö —ç–ª–µ–º–µ–Ω—Ç–æ–≤ —Å–ø–∏—Å–∫–∞, —Ä–∞—Å–ø–æ–ª–æ–∂–µ–Ω–Ω—ã—Ö –º–µ–∂–¥—É –µ–≥–æ –º–∏–Ω–∏–º–∞–ª—å–Ω—ã–º –∏ –º–∞–∫—Å–∏–º–∞–ª—å–Ω—ã–º —ç–ª–µ–º–µ–Ω—Ç–∞–º–∏, –≤–∫–ª—é—á–∞—è –º–∏–Ω–∏–º–∞–ª—å–Ω—ã–π –∏ –º–∞–∫—Å–∏–º–∞–ª—å–Ω—ã–π —ç–ª–µ–º–µ–Ω—Ç—ã.

```python
n=[1,2,3,4,5,6,7,8,4,3]
mx=n.index(max(n))
mn=n.index(min(n))
nn=n[mn:mx+1]
print(sum(nn))
```

–î–∞–Ω —Å–ø–∏—Å–æ–∫. –û–±–Ω—É–ª–∏—Ç—å —ç–ª–µ–º–µ–Ω—Ç—ã —Å–ø–∏—Å–∫–∞, —Ä–∞—Å–ø–æ–ª–æ–∂–µ–Ω–Ω—ã–µ –º–µ–∂–¥—É –µ–≥–æ –º–∏–Ω–∏–º–∞–ª—å–Ω—ã–º –∏ –º–∞–∫—Å–∏–º–∞–ª—å–Ω—ã–º —ç–ª–µ–º–µ–Ω—Ç–∞–º–∏ (–Ω–µ –≤–∫–ª—é—á–∞—è –º–∏–Ω–∏–º–∞–ª—å–Ω—ã–π –∏ –º–∞–∫—Å–∏–º–∞–ª—å–Ω—ã–π —ç–ª–µ–º–µ–Ω—Ç—ã).

```python
n=[1,2,3,4,5,6,7,8,4,3]
mx=n.index(max(n))
mn=n.index(min(n))
for i in range(mn+1,mx):
    n[i]=0
print(n)
```

–î–∞–Ω —Å–ø–∏—Å–æ–∫. –£–≤–µ–ª–∏—á–∏—Ç—å –≤—Å–µ —á–µ—Ç–Ω—ã–µ —á–∏—Å–ª–∞, —Å–æ–¥–µ—Ä–∂–∞—â–∏–µ—Å—è –≤ —Å–ø–∏—Å–∫–µ, –Ω–∞ –∑–∞–¥–∞–Ω–Ω–æ–µ —á–∏—Å–ª–æ. –ï—Å–ª–∏ —á–µ—Ç–Ω—ã–µ —á–∏—Å–ª–∞ –≤ —Å–ø–∏—Å–∫–µ –æ—Ç—Å—É—Ç—Å—Ç–≤—É—é—Ç, —Ç–æ –æ—Å—Ç–∞–≤–∏—Ç—å —Å–ø–∏—Å–æ–∫ –±–µ–∑ –∏–∑–º–µ–Ω–µ–Ω–∏–π.

```python
n=[1,2,3,4,5,6,7,8,4,3]
c=int(input())
for i in n:
    if i%2==0:
        n[n.index(i)]=i+c
print(n)
```

–î–∞–Ω —Å–ø–∏—Å–æ–∫. –û–±–Ω—É–ª–∏—Ç—å –≤—Å–µ –Ω–µ—á–µ—Ç–Ω—ã–µ —á–∏—Å–ª–∞, —Å–æ–¥–µ—Ä–∂–∞—â–∏–µ—Å—è –≤ —Å–ø–∏—Å–∫–µ. –ï—Å–ª–∏ –Ω–µ—á–µ—Ç–Ω—ã–µ —á–∏—Å–ª–∞ –≤ —Å–ø–∏—Å–∫–µ –æ—Ç—Å—É—Ç—Å—Ç–≤—É—é—Ç, —Ç–æ –æ—Å—Ç–∞–≤–∏—Ç—å —Å–ø–∏—Å–æ–∫ –±–µ–∑ –∏–∑–º–µ–Ω–µ–Ω–∏–π.

```python
n=[1,2,3,4,5,6,7,8,4,3]
for i in n:
    if i%2!=0:
        n[n.index(i)]=0
print(n)
```

–î–∞–Ω —Å–ø–∏—Å–æ–∫. –ù–∞–π—Ç–∏ –º–∏–Ω–∏–º–∞–ª—å–Ω—ã–π —ç–ª–µ–º–µ–Ω—Ç –∏–∑ –µ–≥–æ —ç–ª–µ–º–µ–Ω—Ç–æ–≤ —Å —á–µ—Ç–Ω—ã–º–∏ –Ω–æ–º–µ—Ä–∞–º–∏.

```python
n=[9,2,3,4,5,6,7,8,4,3]
m=[x for x in n if (n.index(x))%2==0]
print(min(m))
```

–î–∞–Ω —Å–ø–∏—Å–æ–∫. –ù–∞–π—Ç–∏ –Ω–æ–º–µ—Ä–∞ —Ç–µ—Ö —ç–ª–µ–º–µ–Ω—Ç–æ–≤ —Å–ø–∏—Å–∫–∞, –∫–æ—Ç–æ—Ä—ã–µ –±–æ–ª—å—à–µ —Å–≤–æ–µ–≥–æ –ø—Ä–∞–≤–æ–≥–æ —Å–æ—Å–µ–¥–∞, –∏ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —Ç–∞–∫–∏—Ö —ç–ª–µ–º–µ–Ω—Ç–æ–≤. –ù–∞–π–¥–µ–Ω–Ω—ã–µ –Ω–æ–º–µ—Ä–∞ –≤—ã–≤–æ–¥–∏—Ç—å –≤ –ø–æ—Ä—è–¥–∫–µ –∏—Ö –≤–æ–∑—Ä–∞—Å—Ç–∞–Ω–∏—è.

```python
n=[9,2,3,4,5,6,7,8,4,3]
k=0
nn=[]
for i in range(len(n)-1):
    if n[i]>n[i+1]:
        nn.append(i)
        k+=1
        
print(nn, k)
```

–î–∞–Ω —Å–ø–∏—Å–æ–∫. –ù–∞–π—Ç–∏ –¥–≤–∞ —Å–æ—Å–µ–¥–Ω–∏—Ö —ç–ª–µ–º–µ–Ω—Ç–∞, —Å—É–º–º–∞ –∫–æ—Ç–æ—Ä—ã—Ö –º–∞–∫—Å–∏–º–∞–ª—å–Ω–∞, –∏ –≤—ã–≤–µ—Å—Ç–∏ —ç—Ç–∏ —ç–ª–µ–º–µ–Ω—Ç—ã –≤ –ø–æ—Ä—è–¥–∫–µ –≤–æ–∑—Ä–∞—Å—Ç–∞–Ω–∏—è –∏—Ö –∏–Ω–¥–µ–∫—Å–æ–≤.

```python
n=[9,2,3,4,5,6,7,8,4,3]
s=-1
for i in range(len(n)-1):
    if n[i]+n[i+1]>s:
        nn=[]
        s=n[i]+n[i+1]
        nn.append(n[i])
        nn.append(n[i+1])
                
print(nn)
```

–î–∞–Ω —Å–ø–∏—Å–æ–∫. –ü–æ–º–µ–Ω—è—Ç—å –º–µ—Å—Ç–∞–º–∏ –µ–≥–æ –º–∏–Ω–∏–º–∞–ª—å–Ω—ã–π –∏ –º–∞–∫—Å–∏–º–∞–ª—å–Ω—ã–π —ç–ª–µ–º–µ–Ω—Ç—ã.

```python
n=[1,2,3,4,5,6,7,8,4,3]
mx=n.index(max(n))
mn=n.index(min(n))
mxx=max(n)
n[mx]=min(n)
n[mn]=mxx
print(n)
```
