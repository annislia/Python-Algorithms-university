```python
s='pipipipi'
ss=''
for i in s:
    if i.isalpha(): #если символ - буква
        if i=='z':
            ss+='a'
        elif i=='Z':
            ss+='A'
        else:
            ss+=chr(ord(i)+1) #ord номер буквы с специальрном коде, chr переводит номер обратно в символ
    else:
        ss+=i
print(ss)
```

```python
s='a2b'
for i in range(1, len(s)):
    if s[i-1].isalpha() and s[i].isalpha(): #проверяем что символы буквы
        if s[i]<s[i-1]:
            print(i+1)
            break
else: print('по алфавиту')
```
